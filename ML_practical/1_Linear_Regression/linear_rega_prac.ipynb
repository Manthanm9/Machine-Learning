```python
df = pd.read_csv("/content/gdrive/My Drive/ML_practical/1_Linear_Regression/realest.csv")
df.drop(df.tail(1).index,inplace=True)
df.head(5)
```

```python

df["Space"].median()
```

```python
df["Space"].fillna(df["Space"].median(), inplace = True)
```

```python
space = df[["Space"]]
space
```

```python
price = df[["Price"]]
price
```

```python
#plt.scatter(df.Price, df.Room, 'b.')
plt.scatter( space, price, color = 'red', marker ='+' )
plt.xlabel('Price')
plt.ylabel('Space')
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()

reg.fit(space, price)
```

```python
reg.predict([[820]])
```

```python
reg.score()
```
