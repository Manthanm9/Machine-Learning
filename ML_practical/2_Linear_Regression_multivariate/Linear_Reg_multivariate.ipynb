**Data Preprocessing: Fill NA values with median value**

```python
df.bedrooms.median()
```

```python
df.bedrooms.mean()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
x = df.drop('price', axis = 'columns')
y = df.price
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()


reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```
